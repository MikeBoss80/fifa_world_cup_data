## Carga, Limpieza y Exportación de Datos

**Objetivo:** Cargar los 5 datasets crudos, aplicar limpieza y transformaciones, y exportar los datos procesados a `data/processed/`.

```python
import pandas as pd
import numpy as np
import sys
sys.path.append('..')
from src.data_loader import load_all_data
from src.cleaning import *
from src.visualization import set_style

set_style()
```

## 1. Cargar datos crudos

```python
data = load_all_data()
editions = data['editions']
matches = data['matches']
top_scorers = data['top_scorers']
teams_2026 = data['teams_2026']
fixtures_2026 = data['fixtures_2026']

print('Editions:', editions.shape)
print('Matches:', matches.shape)
print('Top Scorers:', top_scorers.shape)
print('2026 Teams:', teams_2026.shape)
print('2026 Fixtures:', fixtures_2026.shape)
```

## Inspección de datos crudos

```python
print('=== EDITIONS ===')
editions.info()
print('\n=== MATCHES ===')
matches.info()
print('\n=== TOP SCORERS ===')
top_scorers.info()
print('\n=== 2026 TEAMS ===')
teams_2026.info()
print('\n=== 2026 FIXTURES ===')
fixtures_2026.info()
```

```python
print('=== EDITIONS - Head ===')
display(editions.head())
print('\n=== MATCHES - Head ===')
display(matches.head())
print('\n=== TOP SCORERS - Head ===')
display(top_scorers.head())
print('\n=== 2026 TEAMS - Head ===')
display(teams_2026.head())
print('\n=== 2026 FIXTURES - Head ===')
display(fixtures_2026.head())
```

## Aplicar limpieza

```python
editions_clean = clean_editions(editions)
matches_clean = clean_matches(matches)
top_scorers_clean = clean_top_scorers(top_scorers)
teams_2026_clean = clean_2026_teams(teams_2026)
fixtures_2026_clean = clean_2026_fixtures(fixtures_2026)

print('Cleaning complete!')
```

## Exportar datos limpios

```python
editions_clean.to_csv('../data/processed/editions_clean.csv', index=False)
matches_clean.to_csv('../data/processed/matches_clean.csv', index=False)
top_scorers_clean.to_csv('../data/processed/top_scorers_clean.csv', index=False)
teams_2026_clean.to_csv('../data/processed/teams_2026_clean.csv', index=False)
fixtures_2026_clean.to_csv('../data/processed/fixtures_2026_clean.csv', index=False)

print('All cleaned datasets exported to data/processed/')
```

## Verificación final
