```python
import pandas as pd
import numpy as np
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
import seaborn as sns

from src.visualization import set_style, save_figure

set_style()

matches = pd.read_csv('../data/processed/matches_clean.csv')
print('Data loaded successfully')
```

## Análisis 1: Clasificar notas de partidos

```python
noted_matches = matches[matches['notes'].notna() & (matches['notes'] != '')]
print(f'Matches with notes: {len(noted_matches)} out of {len(matches)}')
display(noted_matches[['year', 'team1', 'score1', 'score2', 'team2', 'notes']].head(20))
```

## Análisis 1: Sorpresas por década

```python
matches['decade'] = (matches['year'] // 10) * 10
upsets_by_decade = matches[matches['is_upset'] == True].groupby('decade').size()
display(upsets_by_decade)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x=upsets_by_decade.index, y=upsets_by_decade.values, palette='viridis', ax=ax)
ax.set_title('Number of Upsets by Decade')
ax.set_xlabel('Decade')
ax.set_ylabel('Number of Upsets')
plt.tight_layout()
plt.show()
```
